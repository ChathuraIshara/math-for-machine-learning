```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('bhp.csv')
df
```

```python
df.head()
```

```python
upper=df.price_per_sqft.quantile(0.999)
upper
```

```python
lower=df.price_per_sqft.quantile(0.001)
lower
```

```python
df[df.price_per_sqft>upper].count()
```

```python
df[df.price_per_sqft<lower].count()
```

```python
newDataFrame=df[(df.price_per_sqft>lower) & (df.price_per_sqft<upper)]
newDataFrame
#Here i have created a new data frame by removing outliers using percentile
```

```python
standard_deviation=newDataFrame.price_per_sqft.std()
standard_deviation
#just calcutaed the sdt
```

```python
upper=newDataFrame.price_per_sqft.mean()+(4*standard_deviation)
upper
#valid upper range
```

```python
lower=newDataFrame.price_per_sqft.mean()-(4*standard_deviation)
lower
#valid lower range
```

```python
df_nooutlier1=newDataFrame[(newDataFrame.price_per_sqft>lower) & (newDataFrame.price_per_sqft<upper)]
df_nooutlier1
#new dataset by removing outliers from std
```

```python
newDataFrame['zscore']=(newDataFrame.price_per_sqft-newDataFrame.price_per_sqft.mean())/newDataFrame.price_per_sqft.std()
newDataFrame
#generated zscore column
```

```python
df_nooutlier2=newDataFrame[(newDataFrame.zscore<4) & (newDataFrame.zscore>-4)]
df_nooutlier2
#new dataset by removing outliers from zscores
```
