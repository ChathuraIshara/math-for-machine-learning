```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('movie_revenues.csv')
df
```

```python
df.head()
```

```python
df['revenue_mn']=df['revenue'].apply(lambda x:x/1000000)
df.head()
```

```python
df.head()
```

```python
stadeviation=df.revenue_mn.std()
stadeviation
```

```python
rmean=df.revenue_mn.mean()
rmean
```

```python
def getZscore(x,m,s):
    return (x-m)/s
df['zscore']=df['revenue_mn'].apply(lambda x:getZscore(x,rmean,stadeviation))
df.head(3)
```

```python
df[df['zscore']>3]
```

```python
rmedian=df['revenue_mn'].median()
def getMadHelpValue(x,m):
    return abs((x-m))

df['madhelpvalue']=df['revenue_mn'].apply(lambda x:getMadHelpValue(x,rmedian))
df.head(3)
```

```python
MAD=df['madhelpvalue'].median()
MAD
```

```python
def getModifiedZscore(x,m,M):
    return (0.6745*(x-m)/M)

df['modifiedZscore']=df['revenue_mn'].apply(lambda x:getModifiedZscore(x,rmedian,MAD))
df.head()
```

```python
df[df['modifiedZscore']>3.5]
```
