```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('student.csv')
```

```python
df
```

```python
df.head()
```

```python
sns.histplot(data=df,x='math_score')
plt.show()
```

```python
df.math_score.mean()
```

```python
sns.histplot(data=df,x='math_score',kde=True)
plt.show()
```

```python
sns.histplot(data=df,x='math_score',hue='gender',kde=True)
plt.show()
```

```python
sns.catplot(data=df,x='ethnicity',y='math_score',kind='box')
plt.show()
```

```python
sns.kdeplot(data=df,x='math_score',hue='ethnicity',common_norm=False)
plt.show()
```

```python
sns.kdeplot(data=df,x='math_score',hue='ethnicity',common_norm=False)
```

```python
sns.scatterplot(data=df,x='writing_score',y='reading_score',hue='gender')
plt.show()
```
